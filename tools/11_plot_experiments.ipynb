```python
import json

mean_data = json.load(open('06_mean_data.json', 'r', encoding='utf-8')) + \
            json.load(open('07_mean_data.json', 'r', encoding='utf-8')) + \
            json.load(open('10_mean_data.json', 'r', encoding='utf-8'))
```

```python
import pandas as pd

df = pd.DataFrame(mean_data)

df.to_csv('mean_data.csv', index=False)
```

```python
%matplotlib inline
import statistics
import matplotlib.pyplot as plt
    
x = []
y = []
c = []

for r in mean_data:
    
    x.append(r['false positive n'])
    y.append(r['true positive n'])
    if r['run'] == '06':
        c.append('#ff6961')
    if r['run'] == '07':
        c.append('#A7C7E7')
    if r['run'] == '10':
        c.append('#00FF7F')

plt.scatter(x, y, c=c, alpha=0.5, s=100)
plt.title('Mean Classifier Performance')
#plt.ylim((-1, 30))
#plt.xlim((-1, 25))
plt.xlabel('false positive n')
plt.ylabel('true positive n')

plt.show()
    
x = []
y = []
c = []

for r in mean_data:
    
    if r['n true positive to false positive'] < 100:

        x.append(r['n true positive to false positive'])
        y.append(r['true positive n'])
        if r['run'] == '06':
            c.append('#ff6961')
        if r['run'] == '07':
            c.append('#A7C7E7')
        if r['run'] == '10':
            c.append('#00FF7F')

plt.scatter(x, y, c=c, alpha=0.5, s=100)
plt.title('Mean Classifier Performance')
#plt.ylim((-1, 30))
#plt.xlim((-1, 25))
plt.xlabel('n true positive to false positive')
plt.ylabel('true positive n')

plt.show()
```
